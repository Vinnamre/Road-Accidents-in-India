calculating the avg. of state wise total accidents per year

```python
avg_13 = round(np.mean(data['State/UT-Wise Total Number of Road Accidents during - 2013']),2)
avg_13
```

```python
avg_14 = round(np.mean(data['State/UT-Wise Total Number of Road Accidents during - 2014']),2)
avg_14
```

```python
avg_15 = round(np.mean(data['State/UT-Wise Total Number of Road Accidents during - 2015']),2)
avg_15
```

```python
avg_16 = round(np.mean(data['State/UT-Wise Total Number of Road Accidents during - 2016']),2)
avg_16
```

visualizing the outputs

```python
labels = ['Accidents in 2013', 'Accidents in 2014', 'Accidents in 2015', 'Accidents in 2016']
sizes = [avg_13, avg_14, avg_15, avg_16]
colors = sns.color_palette('dark', len(labels))
```

```python
plt.figure(figsize=(8,8))
plt.pie(sizes, labels= labels, colors= colors, shadow= False, autopct= '%.4f%%')
plt.title('State/UT wise Total number of accidents per year')
plt.axis('equal')
plt.show()
```

It can seen in the chart that the number of accidents per year are on average same. Which indicates that the government is making very less efforts to prevent happening of accidents.
